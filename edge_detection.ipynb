```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the original image
img = cv2.imread('data/coco/val2017/000000000872.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Display original image
plt.figure()
plt.imshow(img)

# Convert to graycsale
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# Blur the image for better edge detection
img_blur = img_gray
#img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)
plt.figure()
plt.imshow(img_blur,cmap="gray")
 
# Sobel Edge Detection
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
# Display Sobel Edge Detection Images
plt.figure()
plt.imshow(sobelx,cmap="gray")
plt.figure()
plt.imshow(sobely,cmap="gray")
plt.figure()
plt.imshow(sobelxy,cmap="gray")

#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img_blur, -1, kernelx)
img_prewitty = cv2.filter2D(img_blur, -1, kernely)
plt.figure()
plt.imshow(img_prewittx,cmap="gray")
plt.figure()
plt.imshow(img_prewitty,cmap="gray")
 
# Canny Edge Detection
edges = cv2.Canny(image=img_blur, threshold1=100, threshold2=200) # Canny Edge Detection
# Display Canny Edge Detection Image
plt.figure()
plt.imshow(edges,cmap="gray")
```
